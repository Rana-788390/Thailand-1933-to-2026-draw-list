```python
# Thai Lottery Automated Data Collector (1990â€“2025)

import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

print("â³ Thai Lottery data automated extraction started...")

records = []

years = range(1990, 2026)
base_url = "https://www.myhora.com/lottery/year-{}.aspx"

for year in years:
    print(f"ğŸ” Processing year {year}")
    url = base_url.format(year)
    try:
        r = requests.get(url, timeout=20)
        if r.status_code == 200:
            soup = BeautifulSoup(r.text, "html.parser")
            blocks = soup.find_all("div", class_="lot-1")

            for b in blocks:
                text = b.get_text(" ").split()
                date = None
                prize = None

                for t in text:
                    if len(t) == 10 and "/" in t:
                        date = t
                    if t.isdigit() and len(t) == 6:
                        prize = t
                        break

                if date and prize:
                    records.append({
                        "Year": year,
                        "Date": date,
                        "First_Prize_6_Digit": prize,
                        "Pangora_First_4": prize[:4]
                    })
        time.sleep(1)
    except Exception as e:
        print("Error:", e)

df = pd.DataFrame(records)
file_name = "Thai_Lottery_1990_2025.xlsx"
df.to_excel(file_name, index=False)

print("âœ… DONE")
print(f"ğŸ“ File created: {file_name}")
```
